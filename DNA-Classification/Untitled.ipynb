```python
#importing neccesary libraries
import numpy as np
import pandas as pd
```

```python
#importing dataset from uci archive
url='http://archive.ics.uci.edu/ml/machine-learning-databases/molecular-biology/promoter-gene-sequences/promoters.data'
names=['Class','id','Sequence']
df=pd.read_csv(url,names=names)
```

```python
#analyzing dataset format
df.iloc[0:5]
```

```python
#dataset have 2 clasee + is promoter and - is non-promoter
classes=df.loc[:,'Class']
classes.describe()
```

```python
#In this section we will remove the tab character '\t' from dataset
sequences=list(df.loc[:,'Sequence'])

dataset={}

for i,seq in enumerate(sequences):
    nucleotides=list(seq)
    nucleotides=[x for x in nucleotides if x != '\t']
    
    nucleotides.append(classes[i])
    dataset[i]=nucleotides
    
print(dataset[0])
```

```python
#converting dataset into dataframe
df=pd.DataFrame(dataset)
df=df.T
df.rename(columns={57:'Class'},inplace=True)
df
```

```python
df.describe()
```

```python
#since dataset is not numerical so we will convert into numerical by using get_dummies()
numerical_df=pd.get_dummies(df)
numerical_df
```

```python
#There is no need for two class coloums beacuse if one is promoter its definitely not non-promoter
df=numerical_df.drop(columns={'Class_-'})
df.rename(columns={'Class_+':'Class'},inplace=True)
df
```

```python
#importing all the algorithms and metricas that we will use
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import KFold,train_test_split, cross_val_score
```

```python
#splitting the dataset into x(input) and y(output)
x=np.array(df.drop(['Class'],1))
y=np.array(df['Class'])

#splitting the dataset into training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#defining metric
scoring='accuracy'

names=['K Nearest Neighbours','Gausian Process', 'Decision Tree','Random Forest','Neural Net','AdaBoost','Naive Bayes'
      'SVM Linear','SVM RBF','SVM Sigmoid']
#initilizing different algorithm correspponding to names
classifier=[
    KNeighborsClassifier(n_neighbors=3),
    GaussianProcessClassifier(1.0*RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5,n_estimators=10,max_features=1),
    MLPClassifier(alpha=1),
    AdaBoostClassifier(),
    GaussianNB(),
    SVC(kernel='linear'),
    SVC(kernel='rbf'),
    SVC(kernel='sigmoid')
]

#making a dictionary
models=zip(names,classifier)

results=[]
names=[]

for name,model in models:
    kfold=KFold(n_splits=10,random_state=1)
    cv_results=cross_val_score(model, x_train,y_train,cv=kfold,scoring=scoring)
    results.append(cv_results)
    names.append(name)
    print('{0} mean:{1} std:{2}'.format(name,cv_results.mean(),cv_results.std()))
    
#from the result we can see the Adaboost is best fit
    
```

```python
#In this section we will check score for testing data
names=['K Nearest Neighbours','Gausian Process', 'Decision Tree','Random Forest','Neural Net','AdaBoost','Naive Bayes'
      'SVM Linear','SVM RBF','SVM Sigmoid']

classifier=[
    KNeighborsClassifier(n_neighbors=3),
    GaussianProcessClassifier(1.0*RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5,n_estimators=10,max_features=1),
    MLPClassifier(alpha=1),
    AdaBoostClassifier(),
    GaussianNB(),
    SVC(kernel='linear'),
    SVC(kernel='rbf'),
    SVC(kernel='sigmoid')
]

models=zip(names,classifier)
for name,model in models:
    print("========================================")
    model.fit(x_train,y_train)
    y_predict=model.predict(x_test)
    print(name)
    print(accuracy_score(y_test,y_predict))
    print(classification_report(y_test,y_predict))
    
#The result on testing data show SVC with RBF kernel is best on prediction
```
